```python
import warnings
warnings.filterwarnings("ignore")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('/Users/mathias/Documents/EM Rates Trading/africa-macro-data-1980-2022.csv')
```

```python
selected_columns = ['Country', 'Year', 'Real GDP growth (annual %)', 'Inflation, consumer prices (annual %)', 'Exports of goods and services (% of GDP)','Imports of goods and services (% of GDP)', 'Gross capital formation (current US$)', 'Gross capital formation (current US$)', 'Current account balance (Net, BoP, cur. US$)', 'General government final consumption expenditure (% of GDP)', 'Household final consumption expenditure  (% of GDP)']
dataset_selected = data[selected_columns]                    
```

```python
selected_country = 'South Africa'
SA_df = dataset_selected[dataset_selected['Country'] == selected_country]
SA_df = SA_df.dropna()
```

```python
SA_df
```

```python
country = SA_df['Country']
year = SA_df['Year']
gdp_growth = SA_df['Real GDP growth (annual %)']
inflation = SA_df['Inflation, consumer prices (annual %)']
exports_gdp = SA_df['Exports of goods and services (% of GDP)']
imports_gdp = SA_df['Imports of goods and services (% of GDP)']
capital_formation_gdp = SA_df['Gross capital formation (current US$)']
current_account_balance = SA_df['Current account balance (Net, BoP, cur. US$)']
govt_consumption_gdp = SA_df['General government final consumption expenditure (% of GDP)']
household_consumption_gdp = SA_df['Household final consumption expenditure  (% of GDP)']
```

```python
SA_df.describe()
```

```python
year = np.array(year)
gdp_growth = np.array(gdp_growth)
inflation = np.array(inflation)
exports_gdp = np.array(exports_gdp)
```

```python
fig1, axs = plt.subplots(2,1, figsize=(15,15,))

axs[0].plot(year, gdp_growth, color='green', label='Real GDP growth (annual %)', linestyle='-', marker='o')
axs[0].set_xlabel('Time (year)')
axs[0].set_ylabel('Real GDP growth (annual %)')
axs[0].set_title('Line Plot: GDP Growth Trends for South Africa')
axs[0].grid(linestyle='--', alpha=0.6)
axs[0].legend()

axs[1].bar(year, gdp_growth, alpha=1, color='red', label='Real GDP growth (annual %)')
axs[1].set_xlabel('Year')
axs[1].set_ylabel('Real GDP growth (annual %)')
axs[1].set_title('Bar Plot: Bar Chart of GDP Growth Trends for South Africa')
axs[1].grid(linestyle='--', alpha=0.6)
axs[1].legend()

axs[1].set_xticks(year)  # Set the tick positions to be the years
axs[1].set_xticklabels(year, rotation=45) 

plt.tight_layout()  # Ensure proper spacing between subplots
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(15,15))

ax.scatter(gdp_growth, inflation, color='green', label='GDP Growth vs Inflation', marker='o')

ax.set_xlabel('Real GDP Growth (annual %)')
ax.set_ylabel('Inflation, consumer prices (annual %)')
ax.set_title('Regression Analysis: Real GDP Growth vs Inflation in South Africa')
ax.grid(True)
ax.legend()

regression_model = np.polyfit(gdp_growth, inflation, 1)
regression_line = np.poly1d(regression_model)

x_range = np.linspace(min(gdp_growth), max(gdp_growth)+5, 100)

ax.plot(x_range, regression_line(x_range), color='red', linestyle='--', label='Regression Line')

plt.legend()
plt.show()
```
